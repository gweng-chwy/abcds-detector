# ABCDs Detector CSV Visualization

Split detector CSV sidecar output into Long-form ABCD and Shorts tables.


```python
import csv, json
from pathlib import Path
root = Path('.')
with (root / 'data/sample_openai_results_check.csv').open(newline='', encoding='utf-8') as f:
    wide_rows = list(csv.DictReader(f))
with (root / 'data/sample_openai_results_check.json').open(encoding='utf-8') as f:
    result = json.load(f)
long_ids = [f['id'] for f in result['assessments'][0]['long_form_abcd_evaluated_features']]
shorts_ids = [f['id'] for f in result['assessments'][0]['shorts_evaluated_features']]
len(wide_rows), len(long_ids), len(shorts_ids)
```

```python
def split_table(feature_ids):
    return [{k: row.get(k, '') for k in ['video_uri', 'brand_name'] + feature_ids} for row in wide_rows]
long_form_table = split_table(long_ids)
shorts_table = split_table(shorts_ids)
long_form_table[0]
```
